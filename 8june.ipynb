```python
class Person:

  def __init__ (self, name, age, ht):
    self.name = name
    self.__age = age
    self.height = ht

  def show_info(self):
    return f"name -> {self.name}, age -> {self.__age}, height -> {self.height}"
```

```python
p1 = Person('jatin', 20, 5.8)
p2 = Person('sachin', 22, 5.9)
```

```python
p1.show_info()
```

```python
from os import name
class Employee:
  def __init__ (self, company, name, salary, job_designation):
    self.company = company
    self.name = name
    self.__salary = salary
    self.job_desig = job_designation
  def anual_sal(self):
    return self.__salary * 12


  def show(self):
    return f"company -> {self.company}, name -> {self.name}, salary -> {self.__salary}, job_desig -> {self.job_desig}"
```

```python
e1 = Employee('wipro', 'abc', 12000, 'HR')
e2 = Employee('wipro', 'xyz', 30000, 'Engineer')
e3 = Employee('wipro', 'pqr', 15000, 'manager')
```

```python
e1.show()
```

```python
# 1
class BankAccount:
    def __init__(self, account_holder, account_number, balance=0):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"₹{amount} deposited successfully.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance! Withdrawal cancelled.")
        else:
            self.balance -= amount
            print(f"₹{amount} withdrawn successfully.")

    def get_balance(self):
        return self.balance

    def __str__(self):
        return (f"Account Holder: {self.account_holder}\n"
                f"Account Number: {self.account_number}\n"
                f"Balance: ₹{self.balance}")

```

```python
account1 = BankAccount("Sanskruti Shingade", "ACC1001", 10000)
account2 = BankAccount("Riya Sharma", "ACC1002", 5000)


account1.deposit(2000)
account1.withdraw(3000)
account1.withdraw(15000)

account2.deposit(1000)
account2.withdraw(2000)
account2.withdraw(6000)
```

```python
print(account1)

print(account2)
```

```python
#2
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        per = self.calculate_percentage()

        if per >= 80:
            return "A"
        elif per >= 60:
            return "B"
        elif per >= 40:
            return "C"
        else:
            return "Fail"

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Percentage:", self.calculate_percentage())
        print("Grade:", self.get_grade())
        print()


students = [
    Student("Amit", 1, [80, 85, 90, 75, 88]),
    Student("Neha", 2, [60, 65, 70, 55, 68]),
    Student("Riya", 3, [35, 40, 30, 45, 38])
]

for s in students:
    s.display()
```

```python
# 3
class Employee:
    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.set_salary(salary)

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_emp_id(self):
        return self.__emp_id

    def set_emp_id(self, emp_id):
        self.__emp_id = emp_id

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary < 10000:
            raise ValueError("Salary must be at least 10000")
        self.__salary = salary

    def give_raise(self, percent):
        self.__salary += self.__salary * percent / 100


emp = Employee("Raj", 101, 20000)
emp.give_raise(10)

print(emp.get_salary())
```

```python
# 4
class SecureVault:
    def __init__(self, pin):
        self.__pin = pin
        self.__data = {}

    def add_entry(self, pin, key, value):
        if pin == self.__pin:
            self.__data[key] = value
        else:
            print("Access Denied")

    def get_entry(self, pin, key):
        if pin == self.__pin:
            return self.__data.get(key)
        else:
            print("Access Denied")


vault = SecureVault("1234")

vault.add_entry("1234", "Email", "abc@gmail.com")
print(vault.get_entry("1234", "Email"))
```

```python
# 5
import math

class Shape:
    def area(self):
        return 0

    def describe(self):
        print("I am a shape")


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def describe(self):
        print("I am a Circle")


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def describe(self):
        print("I am a Rectangle")


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

    def describe(self):
        print("I am a Triangle")


shapes = [Circle(5), Rectangle(4, 6), Triangle(5, 8)]

for s in shapes:
    s.describe()
    print("Area:", s.area())
```

```python
# 6
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")


class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def start_engine(self):
        print(f"{self.make} {self.model}: Engine started")


class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print(f"Silent start - battery at {self.battery_capacity} kWh")


c = Car("Toyota", "Fortuner", 2023, 4)
e = ElectricCar("Tesla", "Model 3", 2024, 75)

c.start_engine()
e.start_engine()
```

```python
# 7
class Animal:
    def sound(self):
        print("Some sound")


class Dog(Animal):
    def sound(self):
        print("Bark")


class Cat(Animal):
    def sound(self):
        print("Meow")


class Cow(Animal):
    def sound(self):
        print("Moo")


def make_all_sounds(animals):
    for animal in animals:
        animal.sound()


animals = [Dog(), Cat(), Cow()]
make_all_sounds(animals)
```

```python
# 8
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

    def multiply(self, a, b, c=1):
        return a * b * c


class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"


calc = Calculator()

print(calc.add(5, 6))
print(calc.add(5, 6, 7))

v1 = Vector(2, 3)
v2 = Vector(4, 5)

print(v1 + v2)
```

```python
# 9
# Write
names = ["Amit", "Riya", "Neha", "Raj", "Pooja"]

with open("students.txt", "w") as file:
    for name in names:
        file.write(name + "\n")

# Read with line numbers
with open("students.txt", "r") as file:
    for i, line in enumerate(file, start=1):
        print(i, line.strip())

# Append
with open("students.txt", "a") as file:
    file.write("Karan\n")
    file.write("Anjali\n")

# Count students
with open("students.txt", "r") as file:
    lines = file.readlines()
    print("Total Students:", len(lines))

# Backup
with open("students.txt", "r") as source:
    content = source.read()

with open("students_backup.txt", "w") as backup:
    backup.write(content)
```
